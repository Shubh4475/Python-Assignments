**Q1.  What is the difference between a function and a method in Python?**

**Ans:** In Python, the terms function and method are both used to refer to callable objects, but they have key differences based on how they are defined and used.

**Function:**

* A function can be referred to as a segment of code that can be defined with the def keyword or in the form of a lambda expression.
* This implies that it can be invoked without the need for an object.
* Usually, functions are defined on a module level or within another function.

**Example of a function:**

```python
def add(a, b):
    return a + b

result = add(3, 5)  # Function call
print(result)  # Output: 8

# In this example, add() is a function that takes two arguments and returns their sum.
```

**Method:**
* A method is a function that is associated with an object (usually an instance of a class) and is called on that object.
* Methods are defined within classes, and they always have at least one parameter (usually self), which refers to the object the method is being called on.
* Methods are called using dot notation on the object (e.g., object.method()).

**Example of a method:**

```python
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(3, 5)  # Method call
print(result)  # Output: 8

# Here, add() is a method of the Calculator class, and it must be called on an instance of Calculator (in this case, calc).
```

**Key Differences:**

**Association:**

* A function is not associated with any object.
* A method is always associated with an object and is called on that object.

**Definition:**

* A function is defined independently using def or lambda.
* A method is defined inside a class and takes self as its first argument (unless it's a static method).

**Usage:**

* A function is called by its name, passing arguments directly (e.g., add(3, 5)).
* A method is called on an object (e.g., calc.add(3, 5)).

To summarize, methods are functions that belong to an object, while functions are standalone blocks of code that can be used anywhere in the program.

**Q2. Explain the concept of function arguments and parameters in Python.**



**Ans:** **Function Parameters:**

* Names given inside a function's definition or signature are parameters. Parameters will be substituted for instead of passing in the values, which are to be given to the function when it is called.
* The parameters specify what is going to be entering the function: the numbers of values expected and often their types as well.

**Example:**

```python
def greet(name, age):  # 'name' and 'age' are parameters
    print(f"Hello {name}, you are {age} years old.")

# Here name and age are parameters. They are variables that will receive the values when the function is called.
```

 **Function Arguments:**
* Arguments are the actual values or data you pass into a function when calling it.
* These values are assigned to the function's parameters during the call.

```python
greet("Alice", 30)  # "Alice" and 30 are arguments

# Here "Alice" and 30 are arguments. They are the actual values passed to the parameters name and age of the greet function.
```

**Types of Function Arguments in Python:**

Python allows different types of arguments, making functions flexible in how they accept input.

**1. Positional Arguments:**
The simplest form of arguments, where values are passed in the order corresponding to the parameters in the function definition.**Example:**

```python
def add(a, b):
    return a + b

result = add(3, 4)  # 3 and 4 are positional arguments
print(result)  # Output: 7

# Here, a receives 3 and b receives 4 because the arguments are passed in the same order as the parameters.
```

**2. Keyword Arguments:**
You can specify arguments by naming the parameter during the function call, allowing them to be passed in any order.**Example:**

```python
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet(age=25, name="Bob")  # 'name' and 'age' are keyword arguments

# Here, name="Bob" and age=25 are keyword arguments. The order of the arguments does not matter, as long as the parameter names are specified.
```

**3. Default Arguments:**
A parameter can have a default value, which is used if no argument is provided for that parameter when the function is called.**Example:**

```python
def greet(name, age=30):
    print(f"Hello {name}, you are {age} years old.")

greet("Charlie")  # Uses the default value for age
greet("David", 40)  # Overrides the default value for age

# In this example, age=30 is a default argument. If age is not provided when the function is called, it defaults to 30.
```

**4. Variable-Length Arguments:**
* Sometimes you may not know in advance how many arguments a function will take. Python allows functions to accept an arbitrary number of arguments using the ***args** (for positional arguments) or ****kwargs** (for keyword arguments).

***args (Positional Variable-Length Arguments)**

* Collects extra positional arguments as a tuple.

```python
def sum_all(*numbers):
    return sum(numbers)

result = sum_all(1, 2, 3, 4)  # 1, 2, 3, and 4 are passed as *numbers
print(result)  # Output: 10

# Here, *numbers collects all the passed arguments into a tuple.
```

****kwargs (Keyword Variable-Length Arguments)**

* Collects extra keyword arguments as a dictionary.

```python
def display_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

display_info(name="Eve", age=22, city="New York")

# Here, **info collects all the passed keyword arguments into a dictionary.
```

**5. Keyword-Only Arguments:**
You can enforce that certain arguments are passed as keyword arguments (i.e., by name) by placing them after a * in the function definition.

```python
def greet(name, *, age):
    print(f"Hello {name}, you are {age} years old.")

greet("Alice", age=30)  # Correct
# greet("Alice", 30)  # Error: missing keyword for 'age'

# Here, age must be passed as a keyword argument.
```

**Q5.  What are iterators in Python and how do they differ from iterables?**

**Ans:** Iterables and iterators are different but somewhat related concepts in Python. To break them down one by one:

**Iterable:** An iterable in Python refers to anything that may produce an iterator. This really encompasses a wide range of objects that can be iterated through the use of a for loop-to include objects that either define an  **__** **iter__()** method or possess a **___** **getitem__()** method, which facilitates iteration.

Common examples of iterables include:

1. Lists ([])

2. Tuples (())

3. Strings ("")

4. Dictionaries ({}) — when iterated, they return keys by default

5. Sets ({})

The key feature of an iterable is that you can obtain an iterator from it using the **iter()** function.

```python
# Example:
my_list = [1, 2, 3]
iterator = iter(my_list)  # my_list is an iterable, and iter() gives us an iterator

for item in my_list:  # we can loop over the iterable directly
    print(item)
```

**Iterator:**
An iterator is an object representing a stream of data and tracks its own current state during an iteration process. It's the result of calling iter() on an iterable. An iterator must implement two methods:

* **__** **iter__()** — returns the iterator object itself (required by the iterable protocol).
* **__** **next__()** - returns the next item in a sequence, or raises a StopIteration exception in the event that there are no more items to return.

In other words, the **iterable** can be iterated over several times, but an **iterator** is consumed once, and after it has been exhausted, it cannot be reused unless reset or reinitialized.

```python
my_list = [1, 2, 3]
iterator = iter(my_list)  # Create an iterator from the iterable

# Using the iterator directly
print(next(iterator))  # Outputs: 1
print(next(iterator))  # Outputs: 2
print(next(iterator))  # Outputs: 3
print(next(iterator))  # Error: Raises StopIteration, because the iterator is exhausted
```

**Key Differences:**

**Iterable:**

* Can be any object that implements the **__** **iter__()** method.
* Can be used multiple times to generate new iterators.
* Examples: Lists, tuples, strings, etc.

**Iterator:**

* An iterator is an object created from an iterable using the **iter()** function.
* It keeps track of the state of iteration and can be consumed once (using **next()**).
* After all elements are iterated over, it raises a **StopIteration** exception.

```python
# Example of both iterable and iterator
my_list = [1, 2, 3]  # Iterable
iterator = iter(my_list)  # Create an iterator from the iterable

# Iterating using the iterator
print(next(iterator))  # Outputs: 1
print(next(iterator))  # Outputs: 2
print(next(iterator))  # Outputs: 3
# next(iterator) would now raise StopIteration since the iterator is exhausted
```

**Q6.  Explain the concept of generators in Python and how they are defined.**

**Ans:** In Python, a generator is said to be a special iterator that allows for lazy evaluation over a sequence of values. This means that values will be generated one by one as we need them, instead of all being computed at once. This can be memory efficient, especially in the case of working with large datasets, as it does not take up memory by storing the entire iterable.

**How Generator works:**
In Python, generators are used when defining functions through the **yield** keyword. A function is said to be a generator function if it uses the **yield** keyword. Upon reaching the **yield** statement, the function invokes a sequence of values lazily through an iterator.

Unlike the typical script that returns an execution-the-generative-function value and terminates, a generator function invokes a yield, then pauses and remembers its place in the function until the next value is requested. This behavior can be described as **"lazy evaluation"**.

Generators can be defined in two main ways:

* Using a Generator Function (with yield):
* Using a Generator Expression (similar to list comprehensions).

**1. Generator Function:**
A generator function uses the **yield** keyword to return a value. Each time yield is executed, the function's state is saved, and the next call to next() resumes from the last yield statement.

Here’s an example of a generator function:


```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the current value of count
        count += 1   # Increment the count

# Using the generator
counter = count_up_to(5)

# Iterating through the generator
for num in counter:
    print(num)  # Outputs: 1, 2, 3, 4, 5
```

In this example, **count_up_to()** is a generator function. The **yield** statement returns values one by one, and the function "pauses" until the next value is requested.

When the **for** loop runs, Python automatically calls **next()** on the generator.
Each time **yield** is encountered, it produces the next number in the sequence.
When there are no more values to yield, the generator stops, and the loop terminates.

**2. Generator Expression:**
A generator expression is similar to a list comprehension but uses parentheses () instead of square brackets []. It is a compact way to define a generator.

```python
squares = (x * x for x in range(5))  # Generator expression

for square in squares:
    print(square)  # Outputs: 0, 1, 4, 9, 16

 # Here, (x * x for x in range(5)) is a generator expression that generates the squares of numbers from 0 to 4.
```

**Some of the key features to note while using generators:**

**Memory efficient:** Generators do not store the entirety of the sequence into memory. They generate values on the fly, hence they are wonderful for manipulating large datasets or infinite sequences.

**Lazy evaluation:** Generators compute values only as needed and differ from lists, which store all elements simultaneously.

**State retention:** When a generator yields a value, it keeps in memory the exact point where the value is yielded(hence keeps track of local variables) so that it can continue from there when next() is called the next time.

**Iteration control:** iterators can suspend their execution and be resumed later. If one has large data to iterate over, or computations that run long, this is very useful.

**Q7.  What are the advantages of using generators over regular functions?**

**Ans:** Generators offer several advantages over regular functions, particularly when it comes to memory efficiency and handling large datasets or continuous streams of data. Here are some key benefits:

**1. Memory Efficiency:** Generators generate outputs one at a time only when requested, rather than generating all outputs once and storing them in memory. This benefits considerably when it comes to working with large sets of data.

```python
# Example: Generating a large sequence without storing in memory
def generate_numbers(max):
    for num in range(max):
        yield num

# Using the generator
numbers = generate_numbers(10**6)
```

**2. Lazy Evaluation:** Lazy evaluation is a term that is often heard. Generators produce values one at a time, yielding during their execution and hence upon request: lazy evaluation. Thus, generators enable a programmer to work with very large sequences and very large-but-finite or even infinite data streams without using or running out of memory.

```python
# Example: Lazy evaluation in generators
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

# Using the infinite generator
gen = infinite_sequence()
for _ in range(5):
    print(next(gen))
```

**3. Simplified Code:**
Generators can simplify your code and make it more readable by abstracting away the iteration logic. This can result in cleaner and more maintainable code.

```python
def generate_numbers(n):
    result = []
    for x in range(n):
        result.append(x)
    return result
```

**4. Performance Enhancement:**
When items are produced on demand, this allows generators to perform better, specially when you might need but do not pull all the data or when handling large volumes of data, in stages.

```python
# Example: Improved performance with generators
def filter_even_numbers(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num

# Using the generator to filter even numbers
even_numbers = filter_even_numbers(range(1000000))
```

**5. Infinite Sequences:**
Generators can be used to create infinite sequences, which is not feasible with lists or other data structures that require all elements to be stored in memory.

```python
# Example: Generating an infinite sequence of Fibonacci numbers
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Using the Fibonacci generator
fib = fibonacci()
for _ in range(10):
    print(next(fib))
```

**6. Pipeline Processing:**
Generators can be linked together to create data pipelines, where each generator processes and passes data to the next one in the sequence. This setup enables efficient and modular data processing.

```python
def generate_numbers():
    for i in range(10):
        yield i

def square(numbers):
    for num in numbers:
        yield num * num

def filter_even(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num

# Combine the generators
numbers = generate_numbers()
squared_numbers = square(numbers)
even_numbers = filter_even(squared_numbers)

for num in even_numbers:
    print(num)
```

Generators are a powerful tool in Python that enable efficient and elegant data processing, especially when dealing with large datasets, infinite sequences, or data pipelines.

Unlike standard functions created with the def keyword, lambda functions can accept any number of arguments but are limited to a single expression. When the function is called, this expression is evaluated and returned.

**Syntax of a Lambda Function:**

                   lambda arguments: expression

Example:

```python
# Defining a lambda function
add = lambda x, y: x + y

# Calling the lambda function
result = add(2, 3)
print(result)
```

The map() function applies a given function to all items in an input list.

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

The filter() function creates a list of elements for which a function returns true.

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]
```

**Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
#To create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms, you need to define a class that implements the iterator protocol by providing both __iter__() and __next__() methods.

class FibonacciIterator:
    def __init__(self, num_terms):
        """Initialize the iterator with the specified number of terms."""
        self.num_terms = num_terms  # Number of terms to generate
        self.current = 0  # Tracks the current term in the sequence
        self.a, self.b = 0, 1  # Initial Fibonacci numbers

    def __iter__(self):
        """Return the iterator object itself."""
        return self

    def __next__(self):
        """Return the next Fibonacci number in the sequence."""
        if self.current >= self.num_terms:
            raise StopIteration  # Stop iteration if we've generated enough terms

        # Return the current Fibonacci number
        fib_number = self.a
        # Update the Fibonacci sequence
        self.a, self.b = self.b, self.a + self.b
        self.current += 1
        return fib_number

fibonacci_sequence = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for number in fibonacci_sequence:
    print(number)
```

**Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_2(max_exponent):
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

for power in powers_of_2(5):
    print(power)
```

```python
# Sample list of tuples
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 4)]

# Sorting the list of tuples by the second element
sorted_list = sorted(tuples_list, key=lambda x: x[1])

print("Sorted list:", sorted_list)
```

**Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 100]

# Using map() to convert the list of Celsius temperatures to Fahrenheit
fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)

# Converting the map object to a list to display the results
fahrenheit_list = list(fahrenheit_temperatures)

print("Temperatures in Celsius:", celsius_temperatures)
print("Temperatures in Fahrenheit:", fahrenheit_list)
```

**Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", result)
```

**Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the** **product of the price per item and the quantity. The product should be increased by 10,- € if the value of the** **order is smaller than 100,00 €.**

**Write a Python program using lambda and map.**

```python
# List of orders
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

# Function to calculate the total price(product of the price per item and the quantity) with the given condition
calculate_total = lambda order: (order[0], order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0))

# Applying the map function to calculate the totals
order_totals = list(map(calculate_total, orders))

# Printing the Output
print(order_totals)
```
